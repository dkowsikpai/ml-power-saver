```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Power.csv')
```

```python
data.hist(figsize = (20, 20))
plt.show()
```

```python
cor = data.corr()
plt.figure(figsize= (12, 9))
sns.heatmap(cor, vmax=0.8, square=True)
plt.show()
```
